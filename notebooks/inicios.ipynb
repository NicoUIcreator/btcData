```python

trace1 = go.Scatter(x=dfBTCDaily.index, y=dfBTCDaily['Price'], mode='markers', name='dfBTCDaily')
trace2 = go.Scatter(x=busquedaGoogle.index, y=busquedaGoogle["Absolute Google Search Volume"], mode='markers', name='busquedaGoogle')
data = [trace1, trace2]


layout = go.Layout(title='Gráfico de dispersión interactivo de df1 y df2',
                   xaxis=dict(title='Etiqueta del eje x'),
                   yaxis=dict(title='Etiqueta del eje y'))


fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

```python



# Crear el gráfico de dispersión
plt.figure(figsize=(8, 6))
plt.scatter(range(len(dfBTCDaily["Price"])), dfBTCDaily["Price"], color='blue', label='Datos 1')
plt.scatter(range(len(busquedaGoogle)), busquedaGoogle["Normalized Value (0-100)"], color='red', label='Datos 2')
plt.xlabel('Índice de datos')
plt.ylabel('Valores')
plt.title('Comparación de dos conjuntos de datos')
plt.legend()
plt.show()

# Crear el gráfico de barras
plt.figure(figsize=(8, 6))
plt.bar(range(len(dfBTCDaily["Price"])), dfBTCDaily["Price"], color='blue', label='Datos 1')
plt.bar(range(len(busquedaGoogle)), busquedaGoogle["Normalized Value (0-100)"], color='red', label='Datos 2', alpha=0.5)
plt.xlabel('Índice de datos')
plt.ylabel('Valores')
plt.title('Comparación de la distribución de dos conjuntos de datos')
plt.legend()
plt.show()
```

```python
googleTrend=pd.read_csv("../data/bitcoin-timeline_Glimpse_Google-TrendsVER.csv")
```

```python
googleTrend.rename(columns={"Time":"Date",
                            "Normalized Value (0-100)":"GoogleValues",
                            "Absolute Google Search Volume":"VolumeSearch"}, inplace=True)
```

```python
googleTrend.set_index("Date", inplace=True)
```

```python
googleTrend.index=pd.to_datetime(googleTrend.index)
```

```python
googleTrend
```

```python
mergeBtcGoogle=pd.merge(dfBTCDaily,googleTrend, on="Date")
```

```python
mergeBtcGoogle
```

```python
mergeBtcGoogle.info()
```

```python
plt.figure(figsize=(7,7))
sns.heatmap(mergeBtcGoogle.corr(),
            vmin=-1,
            vmax=1,
            cmap=sns.diverging_palette(180, 10, s=85, l=25, n=7),
            square=True,
            linewidths=.1,
            annot=True);
```

```python
mergeBtcGoogle.iplot()
```

```python
heatmap = go.Heatmap(z=mergeBtcGoogle["VolumeSearch"],
                     x=mergeBtcGoogle.index,
                     y=mergeBtcGoogle['Price'],
                     colorscale='Viridis')

layout = go.Layout(title='Heatmap del merge', xaxis=dict(title='Date'), yaxis=dict(title='BtcPrice'))

fig = go.Figure(data=[heatmap], layout=layout)

iplot(fig, filename='merge_heatmap.html')
```
