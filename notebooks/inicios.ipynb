```python
sns.set_style("darkgrid")
trace1 = go.Scatter(
    x=m1.index,
    y=m1["MANMM101USA189S"],
    xaxis='x1',
    yaxis='y1',
    name = "M1 a lo largo del tiempo"
)
trace2 = go.Scatter(
    x=busquedaGoogle.index,
    y=busquedaGoogle["Absolute Google Search Volume"],
    xaxis='x2',
    yaxis='y2',
    name = "Volume Search Google"
)
trace3 = go.Scatter(
    x=DolarIndex.index,
    y=DolarIndex["Price"],
    xaxis='x3',
    yaxis='y3',
    name = "Dolar Index",
)
trace4 = go.Scatter(
    x=dfBTCDaily.index,
    y=dfBTCDaily["Price"],
    xaxis='x4',
    yaxis='y4',
    name = "BTC Price"
)
data = [trace1, trace2, trace3, trace4]

layout = go.Layout(
    xaxis=dict(
        domain=[0, 0.45],
        anchor='y1'
    ),
    yaxis=dict(
        domain=[0, 0.45],
        anchor='x1'
    ),
    xaxis2=dict(
        domain=[0.55, 1],
        anchor='y2'
    ),
    xaxis3=dict(
        domain=[0, 0.45],
        anchor='y3'
    ),
    xaxis4=dict(
        domain=[0.55, 1],
        anchor='y4'
    ),
    yaxis2=dict(
        domain=[0, 0.45],
        anchor='x2'
    ),
    yaxis3=dict(
        domain=[0.55, 1],
        anchor='x3'
    ),
    yaxis4=dict(
        domain=[0.55, 1],
        anchor='x4'
    ),
    title = 'EDA sobre Bitcoin, relacion precio y otros indicadores a lo largo de la Historia'
)
fig = go.Figure(data=data, layout=layout)
fig.write_html("../4graficas.html")

iplot(fig)
```
